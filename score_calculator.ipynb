```python
import os
import pandas as pd

def clean_data(df):
    df = df.transpose()
    df = df.reset_index()
    df.columns = ["Player", "Country", "Position", "Games", "Win rate", "KDA", "Avg kills", "Avg deaths", "Avg assists",
                  "CSM",  # Average CS per minute game
                  "GPM",  # Average Gold per minite game
                  "KP%",  # Average Kill participation per game
                  "DMG%",  # Average share of team;s total damage to champions per game
                  "DPM",  # Average damage to champions per minute per game
                  "VSPM",  # Average vision score per minute per game
                  "Avg WPM",  # Average Wards per minute per game
                  "Avg WCPM",  # Average Wards cleared per minute per game
                  "Avg VWPM",  # Averahe Vision wards per minute per game
                  "GD@15",  # Average Gold differential at 15 min
                  "CSD@15",  # Average CS differential at 15 min
                  "XPD@15",  # Average XP differential at 15 min
                  "FB%",  # Firstblood participation (kill or assist)
                  "FB Victim",  # Firstblood victims (%)
                  "Penta Kills",  # Total penta kills
                  "Solo Kills"  # Total solo kills
                 ]
    df.set_index("Player", inplace=True)
    
    # Drop 'Country' and 'Position'
    df.drop('Country', axis=1, inplace=True)
    df.drop('Position', axis=1, inplace=True)
    
    # Fill NaNs in 'Solo Kills' with 0
    if 'Solo Kills' in df.columns:
        df['Solo Kills'].fillna(0, inplace=True)

    return df

roles = ["top", "jg", "mid", "bot", "support"]

df_dict = {}

for role in roles:
    csv_path = os.path.join("data", f"{role}_data.csv")
    df_dict[role] = clean_data(pd.read_csv(csv_path))
```
